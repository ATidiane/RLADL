# Graphs of some TME's results

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(); sns.set_style("whitegrid")
```

## Value Iteration and Policy Iteration

```python
value_0 = pd.read_csv("TME2/csv_results/value_0.csv")
policy_0 = pd.read_csv("TME2/csv_results/policy_0.csv")
value_1 = pd.read_csv("TME2/csv_results/value_1.csv")
policy_1 = pd.read_csv("TME2/csv_results/policy_1.csv")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(value_0.index, value_0.rsum.cumsum(), color='darkblue', label='Value')
plt.plot(policy_0.index, policy_0.rsum.cumsum(), color='darkred', label='policy')
plt.legend()
plt.title("Cumulated Reward GridWorld plan 0")
plt.xlabel("Number of episodes")
plt.ylabel("Reward")
plt.show();
```

```python
plt.figure(figsize=(10, 6))
plt.plot(value_1.index, value_1.rsum.cumsum(), color='darkblue', label='Value')
plt.plot(policy_1.index, policy_1.rsum.cumsum(), color='darkred', label='policy')
plt.legend()
plt.title("Cumulated Reward GridWorld plan 1")
plt.xlabel("Number of episodes")
plt.ylabel("Reward")
plt.show();
```

## DQN

```python
dc = pd.read_csv("TME4/csv_results/dqn_cartpole.csv")
dll = pd.read_csv("TME4/csv_results/dqn_lunarlander.csv")
dntc = pd.read_csv("TME4/csv_results/dqn_no_target_cartpole.csv")
dntll = pd.read_csv("TME4/csv_results/dqn_no_target_lunarlander.csv")
```

### CartPole

```python
plt.figure(figsize=(10, 6))
plt.plot(dc.index, dc.reward, color='darkblue', label='DQN')
plt.plot(dntc.index, dntc.reward, color='darkred', label='DQN-no-target-Network')
plt.legend()
plt.title("Reward on CartPole")
plt.xlabel("Number of episodes")
plt.ylabel("Reward")
plt.show();
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dc.index, dc.reward.cumsum(), color='darkblue', label='DQN')
plt.plot(dntc.index, dntc.reward.cumsum(), color='darkred', label='DQN-no-target-Network')
plt.legend()
plt.title("Cumulated Reward CartPole")
plt.xlabel("Number of episodes")
plt.ylabel("Reward")
plt.show();
```

### LunarLander

```python
plt.figure(figsize=(10, 6))
plt.plot(dll.index, dll.reward.cumsum(), color='darkblue', label='DQN')
plt.plot(dntll.index[:1000], dntll.reward[:1000].cumsum(), color='darkred', label='DQN-no-target-Network')
plt.legend()
plt.title("Reward on LunarLander")
plt.xlabel("Number of episodes")
plt.ylabel("Reward")
plt.show();
```
